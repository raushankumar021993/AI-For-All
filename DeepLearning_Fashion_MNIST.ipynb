```
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```
class myCallbacks(keras.callbacks.Callback):
  def on_epochs_end(self,epoch,logs={}):
    if logs.get('acc')>0.9545:
      print("Cancelling the training process")
      self.model.stop_training = True
    
```

```
fashion_mnist = keras.datasets.fashion_mnist
(train_images,train_labels),(test_images,test_labels) = fashion_mnist.load_data()
```

```
import matplotlib.pyplot as plt
plt.imshow(train_images[0])
#print(train_labels[0])
#print(train_images[0])
```

```
train_images = train_images/255.0
test_images = test_images/255.0
```

```
model = keras.models.Sequential([keras.layers.Flatten(),
                               keras.layers.Dense(128,activation=tf.nn.relu),
                               keras.layers.Dense(10,activation = tf.nn.softmax)])
```

```
model.compile(optimizer='sgd',loss='sparse_categorical_crossentropy')
```

```
callback = myCallbacks()
model.fit(train_images,train_labels,epochs = 10,callbacks = [callback])
```

```
model.predict(test_images,test_labels)
```
